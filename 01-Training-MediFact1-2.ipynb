## Preprocess the Text

```python
import nltk
from nltk.corpus import stopwords
from nltk.tokenize import word_tokenize

# Download the stopwords corpus if not already downloaded
nltk.download('stopwords')
nltk.download('punkt')

def remove_stopwords(text):
    # Tokenize the text
    tokens = word_tokenize(text)

    # Remove stopwords
    stop_words = set(stopwords.words('english'))
    filtered_tokens = [word for word in tokens if word.lower() not in stop_words]

    # Join the filtered tokens back into a string
    filtered_text = ' '.join(filtered_tokens)

    return filtered_text

# Example usage:
original_text = '''patient admitted concomitant diagnosis pseudohyponatremia . Ms. < NAME/ > presented intractable nausea vomiting unclear etiology . Na 134 management included 3 L NS monitoring BMP . 0 patient admitted concomitant diagnosis pseudohyponatremia . 1 Ms. < NAME/ > presented intractable nausea vomiting unclear etiology . 2 Na 134 management included 3 L NS monitoring BMP . patient admitted concomitant diagnosis pseudohyponatremia .'''
filtered_text = remove_stopwords(original_text)
print("Original text:", original_text)
print("Text after removing stop words:", filtered_text)
```

## **MediFact2-Error**
### Training

```python
import pandas as pd

combined_df['Filtered_Text'] = combined_df['Text'].apply(lambda x: remove_stopwords(x))
combined_df['Error Sentence'] = combined_df['Error Sentence'].fillna(combined_df['Filtered_Text'])

data = combined_df[['Filtered_Text','Error Sentence']]

# Split dataset into features (X) and target labels (y)
X = data['Filtered_Text']
y = data['Error Sentence']

```

```python
data
```

```python
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.linear_model import LogisticRegression
from sklearn.pipeline import Pipeline
from sklearn.metrics import classification_report


# Define TF-IDF vectorizer
tfidf_vectorizer = TfidfVectorizer()

# Define logistic regression classifier
logreg_classifier = LogisticRegression()

# Create pipeline
pipeline = Pipeline([('tfidf', tfidf_vectorizer),
                     ('clf', logreg_classifier)])

# Train the model
pipeline.fit(X, y)

# Evaluate the model
y_pred = pipeline.predict(X)
print(classification_report(y, y_pred))

# Example inference
new_paragraph = "A 16-year-old female high school student is brought to the physician by her parents for her repeated behavioral problems at home and school during the past 10 months. Her teachers describe her behavior as uncooperative and disruptive as she persistently refuses to answer questions, insults her teachers, and annoys her classmates on a daily basis. At home, her parents try to address her frequent violations of curfew, but attempts at discussing the issue often result in their daughter losing her temper and screaming at her parents. Her grades have deteriorated over the past year. She has no history of psychiatric illness. On questioning, the patient refuses to answer and frequently disrupts the physicianâ€šÃ„Ã´s conversation with the parents. What is the most likely diagnosis in this patient?"
predicted_answer = pipeline.predict([new_paragraph])[0]
print("Predicted answer:", predicted_answer)
```

```python
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.svm import SVC
from sklearn.pipeline import Pipeline
from sklearn.metrics import classification_report
import numpy as np

# Load your training and validation data
#X_train, y_train = load_training_data()  # You need to define this function
#X_valid, y_valid = load_validation_data()  # You need to define this function

# Define TF-IDF vectorizer
tfidf_vectorizer = TfidfVectorizer()

# Define SVM classifier
svm_classifier = SVC(kernel='linear')

# Create pipeline
pipeline = Pipeline([('tfidf', tfidf_vectorizer),
                     ('clf', svm_classifier)])

# Train the model iteratively over epochs
epochs = 10  # Define the number of epochs
batch_size = 32  # Define the batch size

for epoch in range(epochs):
    print(f"Epoch {epoch + 1}/{epochs}")
    # Shuffle training data
    indices = np.arange(len(X_train))
    np.random.shuffle(indices)
    X_train_shuffled = [X_train[i] for i in indices]
    y_train_shuffled = [y_train[i] for i in indices]

    # Iterate over batches
    for batch_start in range(0, len(X_train), batch_size):
        X_batch = X_train_shuffled[batch_start:batch_start+batch_size]
        y_batch = y_train_shuffled[batch_start:batch_start+batch_size]

        # Fit the pipeline to the current batch
        pipeline.fit(X_batch, y_batch)

    # Evaluate on validation data
    y_valid_pred = pipeline.predict(X_valid)
    print("Validation Report:")
    print(classification_report(y_valid, y_valid_pred))

# Final evaluation on training data
y_train_pred = pipeline.predict(X_train)
print("Final Training Report:")
print(classification_report(y_train, y_train_pred))

# Final evaluation on validation data
y_valid_pred = pipeline.predict(X_valid)
print("Final Validation Report:")
print(classification_report(y_valid, y_valid_pred))

# Example inference
new_paragraph = "A 16-year-old female high school student is brought to the physician by her parents for her repeated behavioral problems at home and school during the past 10 months. Her teachers describe her behavior as uncooperative and disruptive as she persistently refuses to answer questions, insults her teachers, and annoys her classmates on a daily basis. At home, her parents try to address her frequent violations of curfew, but attempts at discussing the issue often result in their daughter losing her temper and screaming at her parents. Her grades have deteriorated over the past year. She has no history of psychiatric illness. On questioning, the patient refuses to answer and frequently disrupts the physicianâ€šÃ„Ã´s conversation with the parents. What is the most likely diagnosis in this patient?"
predicted_answer = pipeline.predict([new_paragraph])[0]
print("Predicted answer:", predicted_answer)
```
